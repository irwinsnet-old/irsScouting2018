```python
# Setting Up Our Environment
import pandas as pd
import numpy as np

# We also need to set our working directory to wherever we put the excel spreadsheet with scouting data.
import os
os.getcwd()
```

```python
# Reading the Excel File
#   NOTE: The Excel file must be located in the working directory (see above)
filename = "test.xlsx"
df = pd.read_excel(filename, "Rankings", index_col = None)
df.head()
```

```python
tasks = ['moveBaseline', 'placeGear', 'shootHighBoiler', 'shootLowBoiler']
df = df[df['task'].isin(tasks)]
df
```

```python
# Now let's drop any row with NaN (standf for Not a Number) in the phase column.
df = df.dropna(axis = 0, subset = ['phase'])
df.head()
```

```python
# Stack and set multi-column index
df = df.set_index(['team', 'phase', 'task'])
df.columns = ['successes', 'attempts']
df = df.stack()
df = df.unstack([1, 2, 3])
df = df.sort_index(axis = 1, level = [0, 1])
df.to_excel("rankings.xlsx", sheet_name = "Rankings")

df.head()
```

```python
placeGear_percent = df[('teleop', 'placeGear', 'successes')] / df['teleop']['placeGear']['attempts']
placeGear_percent.head()
```

```python
#df.insert(0, ('teleop', 'placeGear', 'percent'), placeGear_percent)
df = df.sort_index(axis = 1, level = [0, 1])
df
```

```python
for col in df:
    if col[2] == 'successes':
        phase = col[0]
        task = col[1]
        percent = df[(phase, task, 'successes')] / df[(phase, task, 'attempts')]
        col_index = (phase, task, 'percent')
        df.insert(0, col_index, percent)
        df = df.sort_index(axis = 1, level = [0, 1])
df
```
